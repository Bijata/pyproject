## #.How to export your Titanic pivot table to Excel and then visualize it using Seaborn.

```python
# ‚úÖ Step 1: Create a Pivot Table

pivot = pd.pivot_table(data = train, values = ['Survived'], index = ['Sex'],columns=['Pclass'], aggfunc = ['mean'])

print(pivot)
```

```python
#üì§ Step 2: Export Pivot Table to Excel
pivot.to_excel("titanic_pivot.xlsx")  #This creates a file named titanic_pivot.xlsx in your current working directory.

#You can also specify a full path like:
pivot.to_excel(r"D:\analytics\projects\csv file for python bijata\titanic_pivot.xlsx")
```

```python
#Step 3: Visualize Pivot Table with Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Remove the 'All' row and column if margins=True was used
#pivot_viz = pivot.drop('All', axis=0).drop('All', axis=1)

plt.figure(figsize=(8, 5))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Survival Rate by Sex and Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Sex")
plt.show()
```
