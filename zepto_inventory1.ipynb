```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.types import Integer, String, Float, Date
import re #for convert camel to snake
```

```python
# 1️⃣ Load your Excel file
df = pd.read_excel(r"D:\analytics\sql projects\zepto_v1.xlsx",engine='openpyxl')
```

```python
# 2️⃣ Function to convert camelCase / PascalCase → snake_case
def camel_to_snake(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    s2 = re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1)
    return s2.lower().replace(" ", "_")  # also replace any spaces
```

```python
# 3️⃣ Apply function to all column names
df.columns = [camel_to_snake(col) for col in df.columns]
df.columns
```

```python
# 4️⃣ Define MySQL column types
column_types = {
    "category": String(50),
    "name": String(100),
    "mrp": Float(),
    "discount_percent": Integer(),
    "available_quantity": Integer(),
    "discounted_selling_price": Float(),
    "weight_in_gms": Float(),
    "out_of_stock": Integer(),
    "quantity": Integer()
}
```

```python
# 5️⃣ Connect to MySQL
engine = create_engine("mysql+pymysql://piuser:Pass%40123@localhost:3306/piit")
```

```python
# 6️⃣ Upload DataFrame to MySQL
df.to_sql(
    name="zepto",
    con=engine,
    if_exists="replace",  # replace table if exists
    index=False,
    dtype=column_types
)
```
