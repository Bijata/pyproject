# Drop Unnamed Column

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

# change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace('10-May','5-10')
df.head()
```

#  gender distribution

```python
plt.figure(figsize = (6,6))    #width, height (matplotlib = create 2D charts and plots)
ax = sns.countplot(data = df , x = "Gender")  #(seaborn)
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

```python
# from the above chart we have analyzed that:
# the number of females in the data are more than the number of males
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean', 'WritingScore':'mean'})
gb
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb , annot = True)   #annot = True means provides values
plt.title('relationship between Parents Education & Students Scores')
plt.show()
```

```python
#from the above chart we have concluded that the education of parents have a good( marks difference is almost 11 ) 
#impact on the students score
```

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean', 'WritingScore':'mean'})
gb1
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1 , annot = True)
plt.title('Parent Marital Status & Students Scores')
plt.show()
```

```python
#from the above chart we have concluded that the Parent Marital Status have a negligible (difference of 1marks)
#impact on the students score
```

# distrubution of ethinic groups

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count() 
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()
print(groupA['EthnicGroup'])
print(groupA)

```

```python
l = ["group A",  "group B" , "group C", "group D", "group E"]

mlist = [groupA['EthnicGroup'] , groupB['EthnicGroup'], groupC['EthnicGroup'] , groupD['EthnicGroup'], groupE['EthnicGroup']]
print(mlist)

plt.pie(mlist , labels = l , autopct = "%1.2f%%")
plt.title("Distrubution of Ethinic Groups")
plt.show()
```
